```python
import pandas as pd
import numpy as nup

datos = pd.read_csv('presenciaredes.csv', header=0)#leyendo el archivo csv 
print(datos)

arrbidi= nup .asarray(datos)

#Calculando la diferencia de seguidores del mes de Enero y Junio

ene = int (arrbidi [7][3])#tomando dato del mes de enero del archivo csv
jun = int (arrbidi [7][8])#tomando dato del mes de junio del archivo csv

dif = jun - ene #generando la diferencia de seguidores

pros = nup.abs(dif) 

print("\nDiferencia de seguidores: " + str(pros) + " followers" )

    
#calculando el promedio de crecimiento de seguidores de twitter y facebook

facebook = int (arrbidi[1][3]) + int(arrbidi[1][4]) + int(arrbidi[1][5]) + int(arrbidi[1][6]) + int(arrbidi[1][7]) + int(arrbidi[1][8])

promediof = facebook/6

print("\n Promedio de Crecimiento de Followers Facebook: " + str (promediof) + " %")

twitter = int (arrbidi[8][3]) + int(arrbidi[8][4]) + int(arrbidi[8][5]) + int(arrbidi[8][6]) + int(arrbidi[8][7]) + int(arrbidi[8][8])
promediot = twitter/6

print("\n Promedio de Crecimiento de Followers Twitter: " + str (promediot) + " %")
```
